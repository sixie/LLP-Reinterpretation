```python
list16 = []
for i in range(len(cross_values_r.iloc[1])):
    for n in range(len(cross_values_r)):
        if cross_values_r.iloc[n,i] == 0.016:
            list16.append([i, n])

list30 = []
for i in range(len(cross_values_r.iloc[1])):
    for n in range(len(cross_values_r)):
        if cross_values_r.round(2).iloc[n,i] == 0.030:
            list30.append([i, n])
            
list150 = []
for i in range(len(cross_values_r.iloc[1])):
    for n in range(len(cross_values_r.round(2))):
        if cross_values_r.round(2).iloc[n,i] == 0.15:
            list150.append([i, n])
            
list1500 = []
for i in range(len(cross_values_r.iloc[1])):
    for n in range(len(cross_values_r.round(1))):
        if cross_values_r.round(1).iloc[n,i] == 1.5:
            list1500.append([i, n])
            
list7500 = []
for i in range(len(cross_values_r.iloc[1])):
    for n in range(len(cross_values_r.round(1))):
        if cross_values_r.round(1).iloc[n,i] == 7.5:
            list7500.append([i, n])
```
